```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Extract first 100 rows (binary classification)
sub_X = iris_df.iloc[0:100, [0, 2]].values  # Selecting two features
sub_y = iris_df.iloc[0:100, 4].values  # Class labels

# Correct the label transformation
sub_y = np.where(sub_y == 'Iris-setosa', 0, 1)

# Define perceptron functions
def initialize_weights(n_features, random_state=1):
    rgen = np.random.RandomState(random_state)
    weights = rgen.normal(loc=0.0, scale=0.01, size=n_features)
    bias = np.float64(0.)
    return weights, bias

def net_input(X, weights, bias):
    return np.dot(X, weights) + bias

def predict(X, weights, bias):
    return np.where(net_input(X, weights, bias) >= 0.0, 1, 0)

def train_perceptron(X, y, eta=0.01, n_iter=50, random_state=1):
    weights, bias = initialize_weights(X.shape[1], random_state)
    errors_list = []
    
    for _ in range(n_iter):
        errors = 0
        for xi, target in zip(X, y):
            update = eta * (target - predict(xi, weights, bias))
            weights += update * xi
            bias += update
            errors += int(update != 0.0)
        errors_list.append(errors)
    
    return weights, bias, errors_list

# Train perceptron model
weights, bias, errors = train_perceptron(sub_X, sub_y, eta=0.1, n_iter=10)

# Plot number of updates per epoch
plt.plot(range(1, len(errors) + 1), errors, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Number of updates')
plt.title('Perceptron Learning')
plt.show()
```
